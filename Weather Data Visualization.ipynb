```python
# Weather Data Visualization

# Step 1: Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the Weather Dataset
url = 'weather.csv'
data = pd.read_csv(url)
```

```python
# Step 3: Data Exploration
print("Data Info:")
print(data.info())
```

```python
print("\nData Head:")
print(data.head())
```

```python
# Step 4: Data Preprocessing
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Handle missing values - filling with forward fill method
data.ffill(inplace=True)  # Use ffill directly
```

```python
# Step 5: Plotting Min and Max Temperature Over Time
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['MinTemp'], color='blue', label='Min Temperature (°C)', alpha=0.6)
plt.plot(data['Date'], data['MaxTemp'], color='red', label='Max Temperature (°C)', alpha=0.6)
plt.title('Daily Minimum and Maximum Temperatures Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Step 6: Summary Statistics
print("\nSummary Statistics:")
print(data[['MinTemp', 'MaxTemp']].describe())
```
